```python
#Fazendo os imports

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import datetime
```

```python
df = pd.read_excel("AdventureWorks.xlsx")
df.head(10)
```

```python
#Verificando o tamanho do nosso dataset, e o tipo dos dados

df.shape
df.dtypes
```

```python
#Receita total das vendas

df["Valor Venda"].sum()
```

```python
#Custo total 

df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.sample(5)


#O round arrendonda em duas casas
round(df["custo"].sum(), 2)
```

```python
#Agora vamos buscar o custo

df["Lucro"] = df["Valor Venda"] - df["custo"]
df.sample(5)
```

```python
#Lucro total

round(df["Lucro"].sum(), 2)
```

```python
#Criando uma coluna com total de dias para enviar

df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
df.head()
```

# Agora queremos saber a média do de envio para cada marca, dessa forma torna-se necessário converter aa coluna Tempo _envio em númerica

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head()
```

```python
#Calculando a média de tempo de envio para cada marca

df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#Verificando se temos dados faltantes

df.isnull().sum()
```

# Verificando o lucro por ano de cada marca

```python
#Agrupando por ano e por marca

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resentando o index 

lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Procurando a quantidade de produtos vendidas

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Plotando um gráfico com o total de produtos vendidos

from matplotlib.pyplot import title


df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).plot.barh(title="Total de Produtos Vendidos")
plt.xlabel("Total")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando as vendas apenas de 2009

df_2009 = df[df["Data Venda"].dt.year == 2009]
df.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```
